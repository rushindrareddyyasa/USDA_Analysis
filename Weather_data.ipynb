```python
al_2017['Avg_Temp_C'] = al_2017['Avg Temperature (K)'] - 273.15
```

```python
al_2017.head()
```

```python
daily_county = al_2017.groupby(
    ['FIPS Code','Year','Month','Day']
).agg({
    'Avg_Temp_C': 'mean',
    'Precipitation (kg m**-2)': 'mean',
    'Relative Humidity (%)': 'mean'
}).reset_index()
```

```python
yearly_county = daily_county.groupby(
    ['FIPS Code','Year']
).agg({
    'Avg_Temp_C': 'mean',
    'Precipitation (kg m**-2)': 'sum',
    'Relative Humidity (%)': 'mean'
}).reset_index()
```

```python
yearly_county = yearly_county.rename(columns={
    'FIPS Code': 'FIPS',
    'Avg_Temp_C': 'Avg_Temp_C',
    'Precipitation (kg m**-2)': 'Total_Rain_mm',
    'Relative Humidity (%)': 'Avg_Humidity'
})
```

```python
import os
import pandas as pd

base_year_path = '/content/drive/MyDrive/Crop_Net/WRF-HRRR Computed Dataset/data/2017'

states = os.listdir(base_year_path)

monthly_data_list = []

for state in states:

    state_path = os.path.join(base_year_path, state)

    if not os.path.isdir(state_path):
        continue

    monthly_files = os.listdir(state_path)

    dfs = []

    for file in monthly_files:
        if file.endswith('.csv'):
            df = pd.read_csv(os.path.join(state_path, file))
            dfs.append(df)

    if len(dfs) == 0:
        continue

    state_full_year = pd.concat(dfs, ignore_index=True)

    # Convert temperature to Celsius
    state_full_year['Avg_Temp_C'] = state_full_year['Avg Temperature (K)'] - 273.15

    # 🔹 Grid → Daily County
    daily_county = state_full_year.groupby(
        ['FIPS Code','Year','Month','Day']
    ).agg({
        'Avg_Temp_C':'mean',
        'Precipitation (kg m**-2)':'mean',
        'Relative Humidity (%)':'mean'
    }).reset_index()

    # 🔹 Daily → Monthly County
    monthly_county = daily_county.groupby(
        ['FIPS Code','Year','Month']
    ).agg({
        'Avg_Temp_C':'mean',
        'Precipitation (kg m**-2)':'sum',
        'Relative Humidity (%)':'mean'
    }).reset_index()

    monthly_county = monthly_county.rename(columns={
        'FIPS Code':'FIPS',
        'Precipitation (kg m**-2)':'Total_Rain_mm',
        'Relative Humidity (%)':'Avg_Humidity'
    })

    monthly_data_list.append(monthly_county)

# Combine all states
weather_2017_monthly = pd.concat(monthly_data_list, ignore_index=True)

weather_2017_monthly.head()
```
